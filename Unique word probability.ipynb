```python
dupes = Train['is_duplicate']
```

```python
sent = {}

for i in range(len(allsen)):
    sent[i] = allsen[i]
```

```python
idup = defaultdict(int)
bdup = defaultdict(int)
indup = defaultdict(int)
bndup = defaultdict(int)
```

```python
onecount= defaultdict(int)
bothcount = defaultdict(int)
```

```python
for i in sent.keys():
    for j in sent[i][0]:
        if j in sent[i][1]:
            bothcount[j] += 1
```

```python
keys = list(counts.keys())
for i in keys:
    if counts[i] < 10:
        counts.pop(i)
```

```python
keys = list(bothcount.keys())
for i in keys:
    if bothcount[i] < 10:
        bothcount.pop(i)
```

```python
keys = set(onecount.keys())
for i in keys:
    if onecount[i] < 10:
        onecount.pop(i)
```

```python
for i in sent.keys():
    for j in sent[i][0]:
        if j not in sent[i][1]:
            onecount[j] += 1
            
    for j in sent[i][1]:
        if j not in sent[i][0]:
            onecount[j] += 1
```

```python
indup = defaultdict(int)
idup = defaultdict(int)
keys = set(onecount.keys())
try:
    for i in range(len(sent)):
        for j in sent[i][0]:   
            if j in keys:
                if j not in sent[i][1] and dupes.iloc[i] == 0:
                    indup[j] += 1
                elif j not in sent[i][1] and dupes.iloc[i] == 1:
                    idup[j] += 1

        for j in sent[i][1]:
            if j in keys:
                if j not in sent[i][0] and dupes.iloc[i] == 0:
                    indup[j] += 1
                elif j not in sent[i][0] and dupes.iloc[i] == 1:
                    idup[j] += 1
except KeyboardInterrupt:
    print(i)
    
```

```python
bdup = defaultdict(int)
bndup = defaultdict(int)

keys = list(bothcount.keys())
try:
    for i in range(len(sent)):
        for j in sent[i][0]:
            if j in keys:
                if j in sent[i][1] and dupes.iloc[i] == 0:
                    bndup[j] += 1
                elif j in sent[i][1] and dupes.iloc[i] == 1:
                    bdup[j] += 1

except KeyboardInterrupt:
    print(i)
```
