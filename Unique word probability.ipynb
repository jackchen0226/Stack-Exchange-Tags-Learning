```python
import uwords
import pickle
import pandas as pd
import numpy as np
from keras.preprocessing.text import text_to_word_sequence
from collections import defaultdict
Train = pd.read_csv("QuoraData/train.csv")
```

```python
upkl = open('utokens.pkl', 'rb')
utokens = pickle.load(upkl)
upkl.close()
```

```python
cpkl = open('counts.pkl', 'rb')
counts = pickle.load(cpkl)
cpkl.close()
```

```python
apkl = open('alltokens.pkl', 'rb')
alltokens = pickle.load(apkl)
apkl.close()
```

```python
spkl = open('allsen.pkl', 'rb')
allsen = pickle.load(spkl)
spkl.close()
```

```python
for i in allsen:
    for j in i[0]:
        if j in utokens:
            counts[j] += 1
    for n in i[1]:
        if n in utokens:
            counts[n] += 1
```

```python
dupes = Train['is_duplicate']
```

```python
len(dupes)
```

```python
idup = defaultdict(int)
bdup = defaultdict(int)
indup = defaultdict(int)
bndup = defaultdict(int)
```

```python
#Find way to increase speed
for i in range(len(allsen)):
    for j in counts.keys():
        if j in allsen[i][0] and j not in allsen[i][1] and dupes.iloc[i] == 0:
            indup[j] += 1
        if j not in allsen[i][0] and j in allsen[i][1] and dupes.iloc[i] == 0:
            indup[j] += 1
        if j in allsen[i][0] and j in allsen[i][1] and dupes.iloc[i] == 0:
            bndup[j] += 1
        if j in allsen[i][0] and j not in allsen[i][1] and dupes.iloc[i] == 1:
            idup[j] += 1
        if j not in allsen[i][0] and j in allsen[i][1] and dupes.iloc[i] == 1:
            idup[j] += 1
        if j in allsen[i][0] and j in allsen[i][1] and dupes.iloc[i] == 1:
            bdup[j] += 1
```

```python
keys = list(counts.keys())
for i in keys:
    if counts[i] < 10:
        counts.pop(i)
```
